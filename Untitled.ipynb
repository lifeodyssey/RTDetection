```python
# read WIM output of MODIS for Ariake to estimate HAB by Feng�fs algorithm
from mpl_toolkits.basemap import Basemap
import netCDF4
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from mpl_toolkits.axes_grid1 import make_axes_locatable
from matplotlib.colors import LogNorm

fig=plt.figure()


filename="I:/Nagoya University/Project/HAB/Data/A20191019Ariake.hdf"
print(filename)
nc_file = netCDF4.Dataset(filename, 'r')
print(nc_file)
#nc_var = nc_file.variables.keys()
#print(nc_var)

# set latitude, Longitude of Image
lat  = nc_file.variables['Latitude'][:]
lon  = nc_file.variables['Longitude'][:]
#Read data
DNchl = nc_file.variables['chlor_a-Conv inv-remapped to \'MODIS\''][:]
slope = nc_file.variables['chlor_a-Conv inv-remapped to \'MODIS\''].Slope
intercept = nc_file.variables['chlor_a-Conv inv-remapped to \'MODIS\''].Intercept
DNchl=np.where((DNchl < 0), 255 + DNchl, DNchl)
chl = 10**(DNchl*slope+intercept)
nc_file.close()
#Set Basemap
m = Basemap(projection='mill',
            llcrnrlon=lon.min(),
            urcrnrlon=lon.max(),
            llcrnrlat=lat.min(),
            urcrnrlat=lat.max(),

            resolution='f')

# convert the lat/lon values to x/y projections.
x,y=m(lon,lat)
##TODO 问题出在从这里开始，明天不要跟着老板画图的这个来了，或者弄明白这个imshow
#m = fig.imshow(chl,extent=(lon.min(), lon.max(), lat.min(), lat.max()),
        # interpolation='nearest', cmap='jet',norm=LogNorm(vmin=0.1, vmax=70))
#https://rabernat.github.io/research_computing/intro-to-basemap.html TODO 明天再弄吧 我崩溃了
#fig.tick_params(labelsize=10)

m.drawcoastlines()
# draw a boundary around the map, fill the background.
# this background will end up being the ocean color, since
# the continents will be drawn on top.
m.drawmapboundary()
# fill continents, set lake color same as ocean color.
m.fillcontinents()
m.drawcountries()
m.drawrivers(color='#0000ff')
parallels=np.arange(lat.min(),lat.max(),0.2)
m.drawparallels(parallels,fontsize=10,linewidth=0.25,dashes=[7,15],
                color='k',labels=[1,0,1,1])
##dash：http://bbs.06climate.com/home.php?mod=space&uid=8934&do=blog&id=3284 参考
##linewidth：线的宽度
meridians = np.arange(lon.min(),lon.max(),0.2)
m.drawmeridians(meridians,fontsize=10,dashes=[7,15],
                linewidth=0.3, color='k',labels=[1,1,0,1])
plt.show()
```
